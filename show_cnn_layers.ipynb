```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras


# CHANGE PICTURE 
image_path = "/content/img.png"
layer_name = 'block1_conv1'

# layer_name2 = 'block1_conv1'

# Load the image and preprocess it
img = keras.preprocessing.image.load_img(image_path, target_size=(224, 224))
img_tensor = keras.preprocessing.image.img_to_array(img)
img_tensor = np.expand_dims(img_tensor, axis=0)
img_tensor /= 255.

# Load a pre-trained model and extract the output of a specific layer
model = keras.applications.vgg16.VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
layer_output = model.get_layer(layer_name).output

# Create a model to obtain the feature maps
feature_maps_model = keras.Model(inputs=model.input, outputs=layer_output)
feature_maps = feature_maps_model.predict(img_tensor)

print(feature_maps.shape)
# Plot the feature maps
# plt.figure(figsize=(20, 20))
# for i in range(feature_maps.shape[3]):
#     ax = plt.subplot(8, 8, i+1,)
#     plt.imshow(feature_maps[0, :, :, i], aspect='auto')
#     plt.axis('off')
# plt.tight_layout()
# plt.savefig(f'{layer_name}.png', bbox_inches='tight')
# plt.show()
```

```python
import cv2
from google.colab.patches import cv2_imshow

# Load image
img = cv2.imread("img.png")

# Divide the image into a grid
rows, cols = img.shape[:2]
row_height = rows//3
col_width = cols//3

# Split the image into 9 sub-images
for i in range(3):
    for j in range(3):
        sub_img = img[i*row_height:(i+1)*row_height, j*col_width:(j+1)*col_width]
        cv2_imshow(f"Sub-image ({i}, {j})", sub_img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
